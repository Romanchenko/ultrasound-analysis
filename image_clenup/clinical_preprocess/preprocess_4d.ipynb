## Quick sanity check

```python
import random
import matplotlib.pyplot as plt
from PIL import Image

output_dir = Path(ROOT) / OUTPUT_SUBDIR
pngs = list(output_dir.rglob("*.png"))
print(f"Total PNGs: {len(pngs)}")

sample = random.sample(pngs, min(6, len(pngs)))
fig, axes = plt.subplots(1, len(sample), figsize=(4 * len(sample), 4))
if len(sample) == 1:
    axes = [axes]
for ax, p in zip(axes, sample):
    img = Image.open(p)
    ax.imshow(img, cmap="gray")
    ax.set_title(p.name, fontsize=8)
    ax.axis("off")
plt.tight_layout()
plt.show()
```
